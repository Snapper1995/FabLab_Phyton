# FabLab Phyton Scores

## We used next parametres for having graduation: Attendance, Understanding the subject, Work at the lesson, Related  Questions, Final Assignment and Average performance.	
## We multiplied each parameter by its own coefficient
### Max 40.2, min 0.
#### Score 1 = 0, 2 = 1-5, 3 = 6-15, 4 = 16-35, 5 = 36-40.2

```python
import numpy as np
import csv

with open("/home/netbug/azam/FabLab2.csv", 'r') as f:
    mtrx = list(csv.reader(f, delimiter=","))
    
    iterrows = iter(mtrx)
    next(iterrows)
    for row in iterrows:
        item = float(row[1])*0.1 + float(row[2])*1 + float(row[3])*1 + float(row[4])*0.1 + float(row[5])*1 + float(row[6])*1
        print(row[0], '%.1f' % round(item, 1))
```
